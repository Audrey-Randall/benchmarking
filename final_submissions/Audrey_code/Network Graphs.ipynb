```python
import matplotlib.pyplot as plt
import numpy as np


f = open("local_bandwidth_results.txt", "r")
times = []
bytes_sent = []
bandwidth = []

for line in f:
    line = line.rstrip()
    data = line.split(',')
    times.append(float(data[0]))
    bytes_sent.append(int(data[1]))
print("Original: ", bytes_sent[0:10])

for i, b in enumerate(bytes_sent):
    if i==0:
        continue
    bandwidth.append((bytes_sent[i] - bytes_sent[i-1])/(times[i] - times[i-1]))

# convert to seconds from nanoseconds
for i, b in enumerate(bandwidth):
    bandwidth[i] /= 1000000000

print(bandwidth[0:10])
start = times[0]
for i, t in enumerate(times):
    times[i] -= start
    #print(times[i]*10000)

#bandwidth[269] = bandwidth[270]
#print(bandwidth[265:275])
```

```python
import matplotlib
font = {'size'   : 22}

matplotlib.rc('font', **font)

plt.figure(figsize=(20,10))
plt.plot(times[1:], bandwidth[0:])
plt.xlabel("Time (s)")
plt.ylabel("Bandwidth (GB/sec)")
title = "Remote Bandwidth"
plt.title(title)
plt.show()
```
